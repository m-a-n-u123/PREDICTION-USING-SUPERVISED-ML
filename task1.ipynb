```python
# Importing required libraries
import pandas as pd                                            # pandas is used for data manipulation
import numpy as np                                             # Numpy is used for data manipulation
import matplotlib.pyplot as plt                                # Matplotlib is used for data visualization
from sklearn.model_selection import train_test_split           # Sklearn is a machine learning library used for various algorithm
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
data= pd.read_csv(url)
print("Data imported successfully")
data.head(25)
```

```python
data.info()
```

```python
data.describe()
```

### **let's plot a graph of Hours Studied VS Percentage scored so that it will give us clear idea about data.**

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

**From the graph above, we can clearly see that there is a positive linear relation between the number of hours studied and percentage of score.**

### **Preparing the data**


```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values  
```

### **Split this data into training and test sets. We'll do this by using Scikit-Learn's built-in train_test_split()**

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

### **Training the model and Algorithm**
We have split our data into training and testing sets, and now lets train our model. 

```python
from sklearn.linear_model import LinearRegression
linearRegressor= LinearRegression()
linearRegressor.fit(X_train, y_train)        #Fit linear model.
y_predict= linearRegressor.predict(X_train)  #Predict using the linear model.
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete.")
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

### **Checking the accuracy scores**

```python
print('Test Score')
print(regressor.score(X_test, y_test))
print('Training Score')
print(regressor.score(X_train, y_train))
```

### **Making Predictions**
Now that we have trained our algorithm, it's time to make some predictions.

```python
data= pd.DataFrame({'Actual': y_train,'Predicted': y_predict})
data
```

## **#Let's predict the score for 9.25 hours**


```python
print('Score of student who studied for 9.25 hours a day', regressor.predict([[9.25]]))
```

### **Evaluating the model**

The final step is to evaluate the performance of algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset.

```python
#Checking the efficiency of model
mean_squ_error = mean_squared_error(y_train, y_predict)
mean_abs_error = mean_absolute_error(y_train, y_predict)
print("Mean Squred Error:",mean_squ_error)
print("Mean absolute Error:",mean_abs_error)
```
